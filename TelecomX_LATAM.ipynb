```python
# Columnas a normalizar
columns_to_normalize = ['customer', 'phone', 'internet', 'account']

# Crear una lista para almacenar los DataFrames normalizados
normalized_dfs = []

# Normalizar cada columna y aÃ±adir el resultado a la lista
for col in columns_to_normalize:
    df_telecomx = pd.json_normalize(df_telecomx[col])
    df_telecomx.tail()

# Concatenar los DataFrames normalizados
# Primero, incluimos las columnas originales 'customerID' y 'Churn'
df_telecomx_normalized = pd.concat([df_telecomx[['customerID', 'Churn']] ] + normalized_dfs, axis=1)

# Mostrar las primeras filas del DataFrame normalizado
display(df_telecomx_normalized.head())
```

```python
# Charges.Total  debe ser un cammpo float pero hay valores con cadenas vacias con valores ausentes por lo que no permite pasar a float directamente se debe remplazar antes esos faltantes
import numpy as np
df_telecomx_normalized['Charges.Total'] = df_telecomx_normalized['Charges.Total'].apply(lambda x : x.replace('$','').replace(',','.').replace(' reais','').strip())
#df_telecomx_normalized['Charges.Total '] =
#df_telecomx_normalized['Charges.Total'].astype(np.float64)
```

```python
import numpy as np
# Reemplazar cadenas vacÃ­as con NaN
df_telecomx_normalized['Charges.Total'] = df_telecomx_normalized['Charges.Total'].replace('', np.nan)
# Convertir la columna a tipo numÃ©rico, manejando los NaN
df_telecomx_normalized['Charges.Total'] = df_telecomx_normalized['Charges.Total'].astype(np.float64)
display(df_telecomx_normalized.head())
```
